## 2번 문제

adult dataframe에 존재하는 `'workclass'` column의 missing value를 most frequent value inplacing(Hot Deck Imputation)기법을 활용하여 처리하시오.

## 2번 문제 답안

````python
import pandas as pd
adult_df = pd.read_csv('https://raw.githubusercontent.com/EugeneYoo/practice_file/main/adult.csv')

#데이터셋 컬럼 확인
print(adult_df['workclass'].value_counts(dropna=False))


#최빈값 저장
most_freq = adult_df['workclass'].value_counts(dropna=True).idxmax()
print('\n')
print("most freq value:", most_freq)
print('\n\n')

#최빈값으로 null 값 대체
adult_df['workclass'].fillna(most_freq, inplace = True)

#대체된 결과 확인
print(adult_df['workclass'].value_counts())
````

## 3번 문제

adult dataframe에 존재하는 `'nativeCountry'` column의 missing value를 Missing Value Deletion(row)기법을 활용하여 처리하시오.

## 3번 문제 답안

````python
import pandas as pd

adult_df = pd.read_csv('https://raw.githubusercontent.com/EugeneYoo/practice_file/main/adult.csv')

print("Before deletion: ", len(adult_df))

#Missing Value Deletion, row delte이므로 axis = 0
adult_df.dropna(subset=['nativeCountry'], how = 'any', axis = 0, inplace=True)

print("After deletion: ", len(adult_df))
````

## 5번 문제 답안

### 데이터 읽어오기

````python
import pandas as pd

adult_df = pd.read_csv('https://raw.githubusercontent.com/EugeneYoo/practice_file/main/adult.csv')
````

### 데이터 확인

````python
adult_df['workclass'].value_counts()
````

````python
# 상위 10개 확인
adult_df['workclass'].head(10)
````

### Ordinal_encoder 적용

````python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

# ordinal_encoder 객체 선언
ordinal_encoder = OrdinalEncoder()

# OrdinalEncoder 적용
class_cat_encoded = ordinal_encoder.fit_transform(adult_df['workclass'].to_numpy().reshape(-1, 1))
class_cat_encoded[:10]
````

### 참조 코드

oridinal encoder는 별도의 설정이 없을시, 알파뱃 순서대로 인코딩을 수행합니다.

별도의 설정을 통해 사용자 커스터마이징된 수치로 변환이 가능합니다.

어떤 숫자로 인코딩이 됐는지 편하게 확인하실 수 있도록 별도의 출력코드를 첨부합니다.

아래 결과랑 비교해보세요

```python
# 원래 값
0           State-gov
1    Self-emp-not-inc
2             Private
3             Private
4             Private
5             Private
6             Private
7    Self-emp-not-inc
8             Private
9             Private
Name: workclass, dtype: object

# 변환 값
array([[6.],
       [5.],
       [3.],
       [3.],
       [3.],
       [3.],
       [3.],
       [5.],
       [3.],
       [3.]])
```

````python
# workclass의 값을 별도로 저장함.
value = list(adult_df['workclass'].value_counts().keys())
# 이후 정렬 수행
value = sorted(value)

# key: 총 8개의 컬럼으로 0~7까지 사용
key = list(range(0,len(value)))

# 두 리스트를 기반으로 dict 생성
ordinal_encode = dict(zip(key, value))

# 각 매핑된 값 출력
print(ordinal_encode)
````

## 6번 문제 답안

### 데이터 읽어오기

````python
import pandas as pd

adult_df = pd.read_csv('https://raw.githubusercontent.com/EugeneYoo/practice_file/main/adult.csv')
````

### 데이터 확인

````python
adult_df['workclass'].value_counts()
````

````python
# 상위 10개 확인
adult_df['workclass'].head(10)
````

### One-hot encoder 적용

````python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# OneHotEncoder 선언
onehot_encoder = OneHotEncoder()

# 변환 적용
class_cat_encoded = onehot_encoder.fit_transform(adult_df['workclass'].to_numpy().reshape(-1, 1))
print(class_cat_encoded.toarray()[:10])
````

````python
# 각 순서대로 원 핫 인코딩이 수행됨.
onehot_encoder.categories_
````
