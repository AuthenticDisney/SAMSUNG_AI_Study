### 1번 문제 답안

#### import

````python
from sklearn.datasets import load_wine
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import MinMaxScaler
````

#### 데이터셋만 따로 받아오기

````python
datasets = load_wine()

X = datasets.data

# 각 column 별 분산 확인
df = pd.DataFrame(datasets.data , columns=datasets.feature_names)
print(df.var())

df
````

#### VarianceThreshold 객체 선언 및 스케일링을 적용한 데이터로 분산 학습 수행

````python
selector = VarianceThreshold(threshold=0.04)

selector = selector.fit(MinMaxScaler().fit_transform(X))
````

#### 학습한 variance 확인 및 선택된 feature 출력.


````python
def get_idx(idx_list):
  return_list = []
  for idx in idx_list:
    column_num = int(idx.lstrip('x'))
    return_list.append(column_num)
  return return_list


# 분산이 0.04 이상인 feature 목록을 받아옴
selector_trans = selector.get_feature_names_out()

# 선택된 idx 번호
select_idx= get_idx(selector_trans)
select_names = []

# 해당하는 feature 받아 오기.
for i in select_idx:
  select_names.append(datasets.feature_names[i])

# 선택된 feature 및 각 feature 별 분산 출력
selector_trans= pd.DataFrame(selector.transform(MinMaxScaler().fit_transform(X)), columns=select_names)
print("VarianceTreshold = 0.04\n")
print(selector_trans.var())


selector_trans
````

### 5번 문제 답안

#### Import & load data

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance
from sklearn.datasets import load_iris

datasets = load_iris()

X = datasets.data
y = datasets.target
````

#### Check feature name


````python
f_name = datasets.feature_names

f_name
````

#### Permutation Importance 객체 선언 및 학습 수행


````python
importances = permutation_importance(RandomForestRegressor().fit(X,y), X, y,n_repeats=10, random_state=42).importances_mean
````

#### Feature importance 순서대로 정렬

````python
feature_ids = (-importances).argsort()
````

#### 중요한 피처 순서대로 출력

````python
print("Critical feature")
for i in range(0, len(feature_ids)):
  print("{}. {}".format(i+1, f_name[feature_ids[i]]))
````

### 1번 문제 답안

#### Wine dataset

````python
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()
wine_X = wine.data
wine_y = wine.target

wine_feature_name = wine.feature_names
print('Number of targets: ',len(set(wine_y)))

pd.DataFrame(wine_X, columns=wine_feature_name).head(3)
````

#### split data

````python
from sklearn.model_selection import train_test_split
wine_X_train, wine_X_test, wine_y_train, wine_y_test = train_test_split(wine_X, wine_y, random_state=42)
````

#### Create Pipeline

````python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge

pipe = make_pipeline(StandardScaler(),PolynomialFeatures(), Ridge())
````

#### Find proper parameters

````python
from sklearn.model_selection import GridSearchCV

param_grid = {'polynomialfeatures__degree': [i for i in range(1,8)],
              'ridge__alpha': [10**i for i in range(-2,3)]}
grid = GridSearchCV(pipe, param_grid=param_grid).fit(wine_X_train, wine_y_train)
````

````python
print(grid.best_estimator_)
````

#### Result
**Best parameter combination**



```python
Pipeline(steps=[('standardscaler', StandardScaler()),
                ('polynomialfeatures', PolynomialFeatures()),
                ('ridge', Ridge(alpha=100))])
```

### 2번 문제 답안

#### Wine dataset

````python
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()
wine_X = wine.data
wine_y = wine.target

wine_feature_name = wine.feature_names
print('Number of targets: ',len(set(wine_y)))

pd.DataFrame(wine_X, columns=wine_feature_name).head(3)
````

#### split data

````python
from sklearn.model_selection import train_test_split
wine_X_train, wine_X_test, wine_y_train, wine_y_test = train_test_split(wine_X, wine_y, random_state=42)
````

#### linearRegression model

````python
from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()
lin_model.fit(wine_X_train, wine_y_train)
````

#### Evaluation

````python
from sklearn import metrics

predict = lin_model.predict(wine_X_train)
acc = metrics.r2_score(wine_y_train, predict)
print('Train acc(wine): {}'.format(acc))


predict = lin_model.predict(wine_X_test)
acc = metrics.r2_score(wine_y_test, predict)
print('Test acc(wine): {}'.format(acc))
````

#### Dimensionality reduction(PCA)

````python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#정규화
scaler = StandardScaler()
wine_X_scaled = scaler.fit_transform(wine_X)

wine_X_train, wine_X_test, wine_y_train, wine_y_test = train_test_split(wine_X_scaled, wine_y, random_state=42)

#차원 축소
pca = PCA(n_components = 2)
pca_wine_X = pca.fit_transform(wine_X_scaled)

wine_X_train[:3]
````

#### split data

````python
from sklearn.model_selection import train_test_split
wine_X_train, wine_X_test, wine_y_train, wine_y_test = train_test_split(pca_wine_X, wine_y, random_state=42)


wine_X_train[:3]
````

#### linearRegression model

````python
from sklearn.linear_model import LinearRegression

lin_model_pca = LinearRegression()
lin_model_pca.fit(wine_X_train, wine_y_train)
````

#### Evaluation

````python
from sklearn import metrics

predict = lin_model_pca.predict(wine_X_train)
acc = metrics.r2_score(wine_y_train, predict)
print('Train acc(pca): {}'.format(acc))

predict = lin_model_pca.predict(wine_X_test)
acc = metrics.r2_score(wine_y_test, predict)
print('Test acc(pca): {}'.format(acc))
````

  * Train acc(wine): 0.9036531026409027
  * Test acc(wine): 0.8870868505192568

* PCA

  * Test acc(pca): 0.7854052580807733
